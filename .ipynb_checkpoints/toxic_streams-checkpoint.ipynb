```python
%matplotlib inline

import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15, 12)
```

```python
na_rails = gpd.read_file('data/ne_10m_railroads_north_america.geojson')
us_states = gpd.read_file('data/cb_2017_us_state_20m')
```

```python
print(na_rails.crs) #4326
print(us_states.crs) #4269
```

```python
fig, ax = plt.subplots()

base_color = '#f0f0f0'
border_color = base_color
line_color = '#448ee4'

us_states.plot(ax=ax, edgecolor='white', color=base_color)
na_rails.plot(ax=ax, color=line_color, zorder=1);

ax.set(xlim=(-130,-65), ylim=(20,65));
```

```python
us_states.geom_type.unique()
```

```python
# create a single polygon object for clipping
us_poly = us_states.geometry.unary_union
# create a new GeoDataFrame by selecting the na_rails that intersect with the us_poly
us_rails = na_rails[na_rails.geometry.intersects(us_poly)]
```

```python
# Create a single polygon object for clipping
us_poly = us_states.geometry.unary_union
rails_spatial_index = na_rails.sindex

# Create a box for the initial intersection
bbox = us_poly.bounds

# Get a list of id's for each road line that overlaps the bounding box 
# and subset the data to just those lines
sidx = list(rails_spatial_index.intersection(bbox))
rails_sub = na_rails.iloc[sidx]

# Clip the data - with these data
clipped = rails_sub.copy()
clipped['geometry'] = rails_sub.intersection(us_poly)

# clipped layer with no null geometry values
us_rails_cipped_2 = clipped[clipped.geometry.notnull()]
```

```python
fig, ax = plt.subplots()

base_color = '#f0f0f0'
border_color = base_color
line_color = '#448ee4'

us_states.plot(ax=ax, edgecolor='white', color=base_color)
us_rails_cipped_2.plot(ax=ax, color=line_color, zorder=1);

ax.set(xlim=(-130,-65), ylim=(20,65))
```
